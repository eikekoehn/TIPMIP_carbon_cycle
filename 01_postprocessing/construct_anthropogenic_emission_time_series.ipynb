## With this script, we construct the anthropogenic carbon emission time series (given the emission rates of the models)

```python
print('Loading packages...')
import sys
sys.path.append('../00_modules/.')
from import_packages import PackageGetter
globals().update(PackageGetter.import_standard_packages_for_analysis_and_plotting())
globals().update(PackageGetter.import_custom_packages())
```

```python
varia = 'fco2antt'
ref_varia = 'fgco2' # reference variable to have info about time etc.
models = ['EC-Earth3-ESM-1']#['GFDL-ESM2M']#['NorESM2-LM']#['GFDL-ESM2M']#['IPSL-CM6-ESMCO2']#,'GFDL-ESM2M','NorESM2-LM','UKESM1-2','EC-Earth3-ESM-1']#['GISSE2.1-G-CC2'].   #]#, ]#, # 
runs = pruns.get_run_list('tipmip_tier1')#[1:2]#[1:]
#ref_year = 1
root_dir = f'./../01_postprocessed_data/global_time_series'

model_dict = pmods.get_model_dict('all')
run_dict = pruns.get_run_dict()

def get_freq_input(model):
    if model == 'GISSE2.1-G-CC2':
        freq_input = 'yearly'
    else:
        freq_input = 'monthly'
    return freq_input

def get_stat(varia,default=True):
    if default == True:
        if varia in ['fco2antt']:
            stat = 'integral'
        else: # varia in ['co2mass']:
            stat = None
    return stat
```

```python
stat = get_stat(varia)
for mdx, model in enumerate(models):
    freq = get_freq_input(model)
    mgrab = MODELgrabber.get_grabber(model)
    member = mgrab.get_member()

    if model == 'IPSL-CM6-ESMCO2':
        continue

    for rdx, run in enumerate(runs):

        if run == 'esm-up2p0':
            emission_rate = model_dict[model].emission_rate # in PgC yr-1
        elif run in ['esm-up2p0-gwl2p0-50y-dn2p0','esm-up2p0-gwl4p0-50y-dn2p0']:
            emission_rate = -1*model_dict[model].emission_rate
        else:
            emission_rate = 0
            
        print(f'Emission rate is {emission_rate} Pg C yr-1.')

        # get the reference da
        with xr.open_dataset(f'{root_dir}/{ref_varia}/{model}/{run}/{member}/{freq}/global_{stat}/{ref_varia}_{model}_{run}_{member}_global_{stat}.nc') as ref_ds:

            if model == 'GFDL-ESM2M':
                seconds_per_year = 365 * 24 * 3600 # they use a noleap calendar
            else:
                seconds_per_year = 365.25 * 24 * 3600    
            
            Pg_to_kg = 1e12  # 1000^3
        
            ref_da = list(ref_ds.data_vars.values())[0]
        
            da = xr.zeros_like(ref_da) + (emission_rate / seconds_per_year) * Pg_to_kg

            ds = da.to_dataset(name=f"fco2antt_global_{stat}")

            ds[f"fco2antt_global_{stat}"].attrs.update({
                "units": "kg s-1",
                "long_name": "Synthetic global CO2 emission flux",
                "description": f"Constructed from prescribed emission_rate {emission_rate}Pg C yr-1",
                "author": "Eike E. Köhn",
                "grid_cell_fractions": 'none',
            })

            save_dir = f'{root_dir}/{varia}/{model}/{run}/{member}/{freq}/global_{stat}'
            os.makedirs(save_dir,exist_ok=True)
            save_string = f'{save_dir}/{varia}_{model}_{run}_{member}_global_{stat}.nc'
            print(f'... saving under {save_string}')
            ds.to_netcdf(save_string,mode='w')#name=f"{varia}_global_{stat}")
        
```
