```python
import xarray as xr
import matplotlib.pyplot as plt
```

## Comparing areacella

```python
areacella_old = xr.open_dataset('/home/ekoehn/jobs/jupyter/TipESM/carbon_cycle_reversibility/model_grids/areacella_fx_GFDL-ESM2M_historical_r0i0p0.nc')
areacella_new = xr.open_dataset('/projets/TipESM/UBERN/TipESM/GFDL-ESM2M/esm-piControl/r1i1p1f1/fx/areacella/gn/v20250510/areacella_fx_GFDL-ESM2M_esm-piControl_r1i1p1f1_gn.nc')
```

```python
areacella_old
```

```python
areacella_new
```

```python
fig, ax = plt.subplots(1,3,figsize=(15,5))
areacella_old.areacella.plot(ax=ax[0])
areacella_new.areacella.plot(ax=ax[1])
c2 = ax[2].pcolormesh(areacella_new.areacella.values-areacella_old.areacella.values)
plt.colorbar(c2,ax=ax[2])
ax[0].set_title('Old areacella')
ax[1].set_title('New areacella')
ax[2].set_title('New minus old')
plt.tight_layout()
plt.show()
```
